## 출력 결과
* 받는 결과 값에 따른 출력 파트
1. 어떤것을 받을지
2. 어떻게 받을것인지
3. 어떻게 변환 하여( 전처리 여부 ) 
3. 어디에 저장 할 것인지

아래 공식 문서와 결과 값을 보면 , 
> 결과 값에 대한 해석이 나와 있는데 
>> 우리가 필요한 것은 결국은 json 형식의 item/items 키 값에 들어 있는 title, link. description 이다. 


예제의 결과 값을 살펴보면, 

```
"items": [
{"title": "<b>컴퓨터</b> - 나무위키",

"link": "https://namu.wiki/w/%EC%BB%B4%ED%93%A8%ED%84%B0",

"description": "[4] 대형 <b>컴퓨터</b>와 개인용 <b>컴퓨터</b>(데스크톱 <b>컴퓨터</b>, 노트북 <b>컴퓨터</b>, 태블릿 <b>컴퓨터</b>)가 대표적인 <b>컴퓨터</b>이고, 스마트폰 [5] , 스마트 워치를 비롯한 웨어러블 디바이스, 게임기, 임베디드 시스템, 일부 가전제품 및... "},

{"title": "좋은 <b>컴퓨터</b>를 만드는 사람들 - 한성<b>컴퓨터</b>",

"link": "https://www.monsterlabs.co.kr/",

"description": "좋은 <b>컴퓨터</b>를 만드는 사람들!"},.......]

```


처럼 나와 있다. 이것을 이해하고 정보를 어떻게 받을지 작성 해야 한다. 

우선 결과 값의 key 값의 items 안에

```
List [ dict{"title": 내용,"link" : 링크, "description": 본문},
       dict{"title": 내용,"link" : 링크, "description": 본문},
       ...,
       dict{"title": 내용,"link" : 링크, "description": 본문} ] 
       
```
 
형식이기 때문에  items 키값으로 접근한후에, 리스트의 인덱스로 접근, 
그리고 다시 title, link, description 의 키 값으로 접근한다. 

**언제든 변화가 가능 함으로 , API 서비스 가이드를 읽고 코드를 작성한다.**

# 예제 구현

##  웹문서_검색 구현

````python
import os
import sys
import urllib.request
import pandas as pd
import json
import re
````

````python
client_id = "h3xGK8GD0yzuDM_vZml3"
client_secret = "hjd0Z341fP"

query = urllib.parse.quote(input('검색 질의 : ')) # quote 인코딩 해줌. utf-8로
idx = 0 
display = 100
start = 1
end = 1000 # 최대 갯수

web_df = pd.DataFrame(columns=("Title", "Link", "Descriptrion"))

for start_index in range(start, end, display): # 디스플레이 값만큼, 한번에 받아오며, 최대 갯수는 1000개 이기 때문에 이와 같이 for문을 작성한다. 
    
    url = "https://openapi.naver.com/v1/search/webkr?query=" + query \
          + "&display=" + str(display) \
          + "&start=" + str(start_index) # 요청변수를 참조해서 작성 한다. 형식이기 때문에 응용하는 코드이다. 


    request = urllib.request.Request(url) # request 및 인증
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    
    # 정보를 받아 오는 부분.
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    if(rescode==200):
        response_body = response.read()
        response_dict = json.loads(response_body.decode('utf-8')) # 파싱을 위한 json 사용 -> json to dict
        items = response_dict['items'] # 우리에게 필요한 것은 items kye 값에 전부 들어 있다. 
        # 딕셔너리의 정보를 받아서 web_df에 저장하는 과정.
        for item_index in range(0, len(items)): # items 안에는 리스트로 구성 각각의 정보를 리스트의 index로 접근해서 받아온다.
            remove_tag = re.compile('<.*?>') # <br> 등의 표시 전처리
            title = re.sub(remove_tag, '', items[item_index]['title'])
            link = items[item_index]['link']
            description = re.sub(remove_tag, '', items[item_index]['description'])
            web_df.loc[idx] = [title, link, description]
            idx += 1 # row 변환
    else:
        print("Error Code:" + rescode)
        
web_df.head()
````
